<h2>Funzione MENU PRINCIPALE</h2>
<p>
    Rappresenta la base di partenza di tutto il programma.
    una volta avviata questa l'utente avrà la possibilità di scorrere e avviare tutte le funzionalità del programma
    interagendo con il menù tramite input da tastiera.
    <p>
        Alla fine di ogni operazione l'utente potrà decidere se tornare al menù principale o uscire dal programma.
    </p>
</p>

```python
def menu():
    while True:                             # Ciclo while per mantenere il programma nel menu finché l'utente non sceglie di uscire.
        print('\n---MENU PRINCIPALE---\n')  
        
        opzioni_menu = {                    # Dizionario che rappresenta le opzioni del menu principale
            '1': 'Aggiungi nuovo viaggio',
            '2': 'Visualizza destinazioni',
            '3': 'Cerca destinazioni',
            '4': 'Filtra destinazioni',
            '5': 'Statistiche',
            '6': 'Modifica viaggio',
            '7': 'Elimina viaggio',
            '0': 'Esci'
        }
        
        for k, v in opzioni_menu.items():       # Stampa tutte le opzioni del menu principale
            print(f"{k}. {v.upper()}")

        scelta = input("Scegli un'opzione, digita il numero corrispondente: ")       # Chiede all'utente di scegliere un'opzione
        
        while scelta not in opzioni_menu.keys():        # Valida l'input dell'utente, richiedendo un'opzione valida in caso di errore
            scelta = input("Opzione non valida. digita il numero corrispondente: ")
        if scelta == '1':                           # Gestisce le diverse opzioni del menu principale
            aggiungi_nuovo_viaggio_da_input()       # Funzione per aggiungere un nuovo viaggio
        elif scelta == '2':
            visualizza_destinazioni()               # Funzione per visualizzare tutte le destinazioni
        elif scelta == '3':
            cerca_destinazioni()                    # Funzione per cercare destinazioni in base a criteri specifici
        elif scelta == '4':      
            filtra_destinazioni()                   # Funzione per filtrare le destinazioni in base a parametri multipli
        elif scelta == '5':
            statistiche()                           # Funzione per visualizzare statistiche sui viaggi
        elif scelta == '6':
            modifica_viaggio()
        elif scelta == '7':
            elimina_viaggio()
        elif scelta == '0':
            print("\n---USCITA---")                 
            break                                   # Termina il programma 
```

<h2>Funzione AGGIUNGI NUOVO VIAGGIO</h2>
<p>
    La funzione permette di aggiungere un nuovo viaggio al listino. 
    E' stato previsto l'accesso alla funzionalità sia per via manuale (richiamando la funzione e dando tutti i
    valori del dizionario come parametro) oppure tramite input da tastiera (che è quella che viene richiamata tramite il menù principale)
</p>
<p>
    Una volta inseriti i dati del nuovo viaggio questo viene salvato nel listino, viene visualizzato un messaggio di conferma
    e viene chiesto se si desidera aggiungerne un altro o si torna in automatico al menù principale.
</p>

```python
def aggiungi_nuovo_viaggio(destinazione, paese, continente, tipo_destinazione, periodo_migliore):   # Funzione per aggiungere un nuovo viaggio
    nuovo_viaggio=viaggio.copy()
    nuovo_viaggio['Destinazione'] = destinazione.title()           
    nuovo_viaggio['Paese'] = paese.title()
    nuovo_viaggio['Continente'] = continente.title()
    nuovo_viaggio['Tipo destinazione'] = tipo_destinazione.title()
    nuovo_viaggio['Periodo migliore'] = [stagione.title() for stagione in periodo_migliore]      
    
    lista_viaggi.append(nuovo_viaggio)                         # Aggiunge 'nuovo_viaggio' a'lista_viaggi'
    print("\n---VIAGGIO AGGIUNTO CORRETTAMENTE---")           
    print(f'\nViaggio aggiunto: {lista_viaggi[-1]}')           # Mostra l'ultimo viaggio aggiunto

def aggiungi_nuovo_viaggio_da_input():                          # Funzione per raccogliere i dati dell'utente per un nuovo viaggio
    print('\n---AGGIUNGI NUOVO VIAGGIO:---\n')
    destinazione = input("Inserisci la destinazione: (o '0' per uscire) ").title() 
    if destinazione == '0':                                     
        print("\n---USCITA---")
        return
    print(destinazione)

    paese = input("Inserisci il paese: (o '0' per uscire) ").title() 
    if paese == '0':                                            
        print("\n---USCITA---")
        return
    print(paese)

    continente = input("Inserisci il continente: (o '0' per uscire) ").title()  
    if continente == '0':                                       
        print("\n---USCITA---")
        return
    print(continente)

    tipo_destinazione = input("Inserisci il tipo di destinazione (Città | Mare | Montagna): (o '0' per uscire) ").title()  
    while tipo_destinazione not in ['Città', 'Mare', 'Montagna', '0', '']:   # Verifica che il tipo di destinazione sia valido (Città, Mare, Montagna)
        print("Tipo destinazione non valido.")                  
        tipo_destinazione = input("Inserisci il tipo di destinazione (Città | Mare | Montagna): (o '0' per uscire) ").title()
    if tipo_destinazione == '0':                                
        print("\n---USCITA---")
        return
    print(tipo_destinazione)

    periodo_migliore = [p.strip() for p in input("Inserisci il periodo migliore (Primavera, Estate, Autunno, Inverno): (o '0' per uscire) ").title().split(',')]
    for i in range(len(periodo_migliore)):             # Verifica che ogni periodo inserito sia valido (Primavera, Estate, Autunno, Inverno)
        while periodo_migliore[i] not in ['Primavera', 'Estate', 'Autunno', 'Inverno', '0','']:
            periodo_migliore = [p.strip() for p in input("Inserisci il periodo migliore (Primavera, Estate, Autunno, Inverno): (o '0' per uscire) ").title().split(',')]
        if '0' in periodo_migliore:                                 
            print("\n---USCITA---")
            return
    print(periodo_migliore)

    aggiungi_nuovo_viaggio(destinazione, paese, continente, tipo_destinazione, periodo_migliore)      
    aggiungi_nuovo_viaggio_da_input() if input("\nVUOI AGGIUNGERE UN ALTRO VIAGGIO? (sì/no)\n : ").strip().lower() in ['yes', 'sì', 'si'] else print("\n---USCITA---")       # Chiede all'utente se vuole aggiungere un altro viaggio


    
```

<h2>Funzione VISUALIZZA DESTINAZIONI</h2>

    I viaggi vengono ordinati in base alla destinazione per via alfabetica.

```python

def visualizza_destinazioni():                              # Funzione per visualizzare la lista delle destinazioni
    print("\n---LISTA VIAGGI---\n")                       
    
    lista_viaggi.sort(key=lambda x: x['Destinazione'])      # Ordina 'lista_viaggi' alfabeticamente in base al campo 'Destinazione'
    
    for viaggio in lista_viaggi:                            
        print(f"Destinazione: {viaggio['Destinazione']}, "
              f"Paese: {viaggio['Paese']}, Continente: {viaggio['Continente']}, "
              f"Tipo destinazione: {viaggio['Tipo destinazione']}, "
              f"Periodo migliore: {', '.join(viaggio['Periodo migliore'])}\n")

    
```

<h2>Funzione CERCA DESTINAZIONI</h2>
<p>
    La funzione è stata pensata per dare l'opportunità all'utente di raggruppare le destinazioni in base a:
    <ul>
        <li>Paese</li>
        <li>Continente</li>
        <li>Tipo destinazione</li>
        <li>Periodo migliore</li>
    </ul>
</p>
<p>
    L'utente sceglie il criterio di ricerca e il valore da cercare e il sistema restituisce tutti i viaggi che rispettano
    il criterio scelto.
    Alla fine può decidere se effettuare un'altra ricerca o torna automaticamente al menù principale dalla quale potrà poi
    uscire dal programma.
</p>

```python
def cerca_destinazioni():                  
    def ricerca_base(chiave):           # Funzione interna per eseguire una ricerca basata su una chiave specifica
        valori_unici = set()            # Insieme per raccogliere valori unici relativi alla chiave
        for v in lista_viaggi:          
            if chiave in v:             
                valore = v[chiave]
                if isinstance(valore, list):  # Se il valore è una lista (esempio, periodo migliore)
                    valori_unici.update(valore)  # Aggiunge tutti gli elementi della lista
                else:
                    valori_unici.add(valore)  
        valori_unici = sorted(valori_unici)  
          
        print(f"\n---VALORI UNICI DI {chiave.upper()}---\n")       # Stampa i valori unici trovati per la chiave
        for valore in valori_unici:
            print(f'- {valore}')
        seleziona_viaggio(chiave)       # Passa alla selezione del viaggio basato sul valore scelto

    def seleziona_viaggio(chiave):      # Funzione per selezionare un valore specifico basato sulla chiave 
        scelta = input(f"Inserisci {chiave} scelta ('0' per uscire): ").title()      # Richiede all'utente di inserire un valore specifico per la chiave
        if scelta=='0':
            return
        print(f'\n--- VIAGGI PER {scelta}:---\n')
        scelta_viaggio(lista_viaggi, scelta, chiave)        # Cerca i viaggi che corrispondono alla scelta
    
    def scelta_viaggio(lista, scelta, chiave):          # Funzione per filtrare e stampare viaggi basati sulla chiave e sul valore scelto
        risultati_trovati = False                       # Flag per verificare se ci sono risultati
        for viaggio in lista:
            if isinstance(viaggio[chiave], list):       # Se il valore della chiave è una lista
                if scelta in viaggio[chiave]:           # Controlla se la scelta è nella lista
                    print(viaggio)
                    risultati_trovati = True
            else:                                       # Se il valore della chiave è un singolo elemento
                if viaggio[chiave] == scelta:           # Confronta il valore con la scelta
                    print(viaggio)
                    risultati_trovati = True
        
        if not risultati_trovati:
            print(f"Nessun viaggio trovato per {scelta}.")  # Messaggio se non ci sono risultati
    
        # Stampa il menu di ricerca per selezionare il criterio di ricerca
    print("\n---RICERCA DESTINAZIONI PER---\n")     
    opzioni_ricerca = {
        '1': 'Paese',                   
        '2': 'Continente',              
        '3': 'Tipo destinazione',       
        '4': 'Periodo migliore',        
        '0': 'Esci'                     
    }
            
    for k, v in opzioni_ricerca.items():        
        print(f"{k}. {v.upper()}")
    
    scelta = input("Scegli un'opzione: ")           # Richiede all'utente di scegliere un'opzione
    while scelta not in opzioni_ricerca.keys():     # Controlla che la scelta sia valida
        scelta = input("Opzione non valida. Scegli un'altra opzione: ")
    
    if scelta == '0':                          
        print("\n---USCITA---")
        return    
    else:
        chiave = opzioni_ricerca[scelta]        # Determina la chiave corrispondente all'opzione scelta
        ricerca_base(chiave)                    # Avvia la ricerca basata sulla chiave scelta
    
    # Chiede all'utente se vuole effettuare un'altra ricerca
    cerca_destinazioni() if input("\nVUOI CERCARE ALTRE DESTINAZIONI? (si/no)\n : ").strip().lower() in ['yes', 'sì', 'si'] else print("\n---USCITA---")
```

        uscire dal programma.</p>
</p>

```python
def filtra_destinazioni():
    filtro_viaggio = {}  # Dizionario per raccogliere i filtri inseriti dall'utente

    # Chiede all'utente di inserire un valore per ciascuna chiave nel dizionario "viaggio"
    for chiave in viaggio:                  # Itera sulle chiavi presenti nel dizionario "viaggio"
        valore = input(f"Inserisci {chiave} o lascia vuoto: ").title()          # Richiede input per la chiave
        if valore.strip():                  # Se l'utente ha inserito un valore (non vuoto)
            filtro_viaggio[chiave] = valore             # Aggiunge il valore al filtro
        else:
            filtro_viaggio[chiave] = None           # Imposta None per ignorare questa chiave nel filtro
  
    filtro_viaggio = {k: v for k, v in filtro_viaggio.items() if v is not None}         # Rimuove dal dizionario filtro le chiavi che hanno valore None

    print(f"\n---HAI FILTRATO PER:---\n {filtro_viaggio}\n")        # Mostra i filtri definiti dall'utente
    
    risultati = []  # Lista per raccogliere i risultati che soddisfano i criteri di filtro

    for destinazione in lista_viaggi:           # Itera su ogni destinazione nella lista dei viaggi per applicare i filtri
        # Controlla che tutti i filtri impostati dall'utente siano soddisfatti
        if all(filtro_viaggio[key] in destinazione[key] for key in filtro_viaggio if filtro_viaggio[key]):
            risultati.append(destinazione)  # Aggiunge la destinazione ai risultati se soddisfa i filtri
    
    if risultati:           # Stampa i risultati trovati, oppure informa l'utente che non ci sono risultati
        print("\n---VIAGGI TROVATI---\n")
        for v in risultati:
            print(v)  
    else:
        print("\n---NESSUN VIAGGIO TROVATO---\n")
    
    # Chiede all'utente se vuole applicare un altro filtro
    filtra_destinazioni() if input("\nVUOI FILTRARE ALTRE DESTINAZIONI? (sì/no)\n : ").strip().lower() in ['yes', 'sì', 'si'] else print("\n---USCITA---")
```

<h2>Funzioni MODIFICA ed ELIMINA VIAGGIO</h2>
<p>
    Le due funzioni sono strutturate in modo simile e permettono intuitivamente all'utente di modificare
    o eliminare viaggi dalla lista.
    Sono interamente accessibili e navigabili da menù principale tramite input da tastiera
</p>
<p>
    N.B:    Sarebbero potute essere accorpate in un unica funzione differenziandole poi nella fase di
            modifica\rimozione ma non avrei risparmiato molte righe di codice quindi ho preferite tenerle
            ben distinte per una questione di ordine personale.
</p>

```python
def modifica_viaggio():
    print('\n--- MODIFICA VIAGGIO ---')
    while True:
        print('\nQUALI VIAGGI VUOI MODIFICARE?')
        visualizza_destinazioni()
        destinazione = input("\nInserisci la destinazione del viaggio da modificare: ('0' per uscire) ").title()
        if destinazione=='0':
            print("\n--- USCITA ---")
            break

        # Cerca il viaggio nella lista
        viaggio_trovato = False
        for val in lista_viaggi:
            if val['Destinazione'] == destinazione:
                viaggio_trovato = True
                print(f'\nModifica i dettagli per il viaggio a {destinazione}.')
                print('Inserisci i nuovi valori o premi INVIO per mantenere il valore attuale.\n')

                for k, v in val.items():
                    if isinstance(v, list):  # Se il valore è una lista
                        attuale = ", ".join(v)
                        nuovo_valore = input(f'{k} (attuale: {attuale}): ').strip().title()
                        if nuovo_valore:
                            val[k] = [item.strip().title() for item in nuovo_valore.split(',')]
                    else:  # Per altri tipi di valori
                        nuovo_valore = input(f'{k} (attuale: {v}): ').strip().title()
                        if nuovo_valore:
                            val[k] = nuovo_valore

                print('\nViaggio modificato con successo:')
                print(val)
                break

        # Se il viaggio non è stato trovato
        if not viaggio_trovato:
            print(f'\n{destinazione} non è nella lista dei viaggi.')
            risposta = input(f'Vuoi aggiungere {destinazione} alla lista? (sì/no): ').strip().lower()
            if risposta in ['sì', 'si', 'yes']:
                aggiungi_nuovo_viaggio_da_input()
            else:
                print('Destinazione non aggiunta.')

        # Chiedi se l'utente vuole modificare altro
        altra_modifica = input("\nVUOI MODIFICARE ALTRO? (sì/no): ").strip().lower()
        if altra_modifica not in ['sì', 'si', 'yes']:
            print("\n--- USCITA ---")
            break
```

```python
def elimina_viaggio():
    while True:
        print('\n---ELIMINA VIAGGIO:---')
        print('\nQUALE VIAGGIO VUOI ELIMINARE?:')
        visualizza_destinazioni()
        destinazione=input("\nInserisci la destinazione del viaggio da eliminare: ('0' per uscire) ").title()
        if destinazione=='0':
            print("\n--- USCITA ---")
            break
        
        viaggio_trovato= False
        for v in lista_viaggi:
            if v['Destinazione']==destinazione:
                viaggio_trovato= True
                print(f'\nHai selezionato:\n {v}')
                scelta=input(f'Confermi eliminazione di {destinazione}?: (sì/no)')
                if scelta in ['si','sì','yes']:
                    lista_viaggi.remove(v)
                    print(f'\nHai eliminato {destinazione} dalla lista viaggi')
                else:
                    print('Viaggio non eliminato')

        if not viaggio_trovato:
            print(f'\n{destinazione} non è nella lista viaggi')
            risposta = input(f'Vuoi aggiungere {destinazione} alla lista? (sì/no): ').strip().lower()
            if risposta in ['sì', 'si', 'yes']:
                aggiungi_nuovo_viaggio_da_input()
            else:
                print('Destinazione non aggiunta.')

        altra_modifica = input("\nVUOI ELIMINARE ALTRO? (sì/no): ").strip().lower()
        if altra_modifica not in ['sì', 'si', 'yes']:
            print("\n--- USCITA ---")
            break
```

<h2>Funzione STATISTICHE</h2>
<p>
    La sezione statistiche si basa su un'analisi dei viaggi presenti nel listino.
    Le statistiche che vengono visualizzate sono:
    <ul>
        <li>Numero totale di viaggi presenti nel listino</li>
        <li>Numero di viaggi per continente</li>
        <li>Numero di viaggi per paese</li>
        <li>Numero di viaggi per tipo destinazione</li>
        <li>Numero di viaggi per periodo migliore</li>
    </ul>
    <p>
        Si basano tutte sullo stesso principio di funzionamento dettato dalla funzione conta_destinazioni che 
        prende come parametri i dati restituiti dalla funzione statistiche tramite interazione con l'utente
    </p>
    <p>Alla fine può decidere se visualizzare altre statistiche o si torna automaticamente al menù principale.</p>
</p>


```python
def conta_destinazioni(chiave, valore):
    valore = valore.title()  
    conteggio = 0  # Inizializza il contatore a zero
    
    # Itera su ogni destinazione nella lista dei viaggi
    for destinazione in lista_viaggi:
        if isinstance(destinazione.get(chiave), list):              # Controlla se il valore associato alla chiave è una lista
            # Se lo è, controlla se il valore è presente nella lista
            if valore in [v.title() for v in destinazione[chiave]]:  
                conteggio += 1                              # Incrementa il conteggio se il valore è trovato
        elif destinazione.get(chiave, '').title() == valore:    # Se il valore associato alla chiave non è una lista (ad esempio, è una stringa)
            conteggio += 1                                      # Incrementa il conteggio se il valore corrisponde esattamente
    return conteggio              # Restituisce il numero di destinazioni che soddisfano il criterio

def statistiche():
    print("\n---STATISTICHE---\n")              # Intestazione per la sezione delle statistiche
 
    statistiche_dict = {                        # Dizionario con le opzioni disponibili per le statistiche
        '1': 'Numero totale di destinazioni',  
        '2': 'Numero di destinazioni per paese',
        '3': 'Numero di destinazioni per continente',
        '4': 'Numero di destinazioni per tipo',
        '5': 'Numero di destinazioni per periodo',
        '0': 'Esci'
    }

    for k, v in statistiche_dict.items():       # Mostra tutte le opzioni statistiche disponibili
        print(f"{k}. {v.upper()}")

    scelta = input("Scegli un'opzione: ")       # Chiede all'utente di scegliere un'opzione
    while scelta not in statistiche_dict.keys():  # Controlla che l'opzione sia valida
        print('\nOpzione non valida')
        scelta = input("Scegli un'altra opzione: ")
    if scelta == '0':                           # Gestisce l'uscita dal menu statistiche
        print("\n---USCITA---")
        return
    elif scelta == '1':             # Mostra il numero totale di destinazioni
        print(f'\nNumero totale di destinazioni: {len(lista_viaggi)}\n')
    elif scelta == '2':         # Mostra il numero di destinazioni per ogni paese
        print('\n---STATISTICHE PER PAESE:---\n')
        paesi = set(destinazione['Paese'].title() for destinazione in lista_viaggi)         # Raccolta di tutti i paesi
        for paese in paesi:
            print(f'{paese}: {conta_destinazioni("Paese", paese)}')         # Conta le destinazioni per paese
    elif scelta == '3':         # Mostra il numero di destinazioni per ogni continente
        print('\n---STATISTICHE PER CONTINENTE:---\n')
        continenti = set(destinazione['Continente'].title() for destinazione in lista_viaggi)       # Raccolta di continenti
        for continente in continenti:
            print(f'{continente}: {conta_destinazioni("Continente", continente)}')          # Conta le destinazioni per continente
    elif scelta == '4':     # Mostra il numero di destinazioni per tipo di destinazione
        print('\n---STATISTICHE PER TIPO DI DESTINAZIONE:---\n')
        tipi = set(destinazione['Tipo destinazione'].title() for destinazione in lista_viaggi)          # Raccolta di tipi di destinazione
        for tipo in tipi:
            print(f'{tipo}: {conta_destinazioni("Tipo destinazione", tipo)}')        # Conta le destinazioni per tipo
    elif scelta == '5':             # Mostra il numero di destinazioni per periodo migliore
        print('\n---STATISTICHE PER PERIODO MIGLIORE:---\n')
        periodi = set(stagione.title() for destinazione in lista_viaggi for stagione in destinazione['Periodo migliore'])
        for periodo in periodi:
            print(f'{periodo}: {conta_destinazioni("Periodo migliore", periodo)}')  # Conta le destinazioni per periodo
    else:           # Gestisce opzioni non riconosciute
        print("\nOpzione non riconosciuta.")
        return

    # Consente all'utente di calcolare un'altra statistica o tornare al menù principale
    statistiche() if input("\nNUOVA STATISTICA? (sì/no) : ").strip().lower() in ['yes', 'sì', 'si'] else print("\n---USCITA---")
```
